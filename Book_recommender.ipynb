```python
books.head()
```

```python
users.head()
```

```python
# Joining Books and user ratings into one table
books_with_ratings = ratings.merge(books, on = 'ISBN')
```

```python
books_with_ratings.head()
```

```python
books_with_ratings.shape
```

```python
popular_df = books_with_ratings.groupby('Book-Title').agg(num_rating=('Book-Rating','count'), 
                                                         avg_rating=('Book-Rating','mean'))
```

```python
popular_df = popular_df.reset_index()
```

```python
popular_df
```

```python
popular_df.sort_values('num_rating',ascending=False)
```

```python
# Popularity is based on the no. of people read the book ('num_rating' > 300) at least 300 people gave the review
# It is based on the rating it got.
popular_df = popular_df[popular_df['num_rating']>300].sort_values('avg_rating',ascending=False)
```

```python
popular_df
```

```python
# Top 50 books
popular_df.head(50)
```

```python
# For model deployment I need Book-title, Author, Image URL
# Merging popular_df and books table for  model deployment
popular_df = popular_df.merge(books, left_on = 'Book-Title',right_on='Book-Title').drop_duplicates('Book-Title')[['Book-Title',
                                                                                       'Book-Author',
                                                                                       'Image-URL-M', 
                                                                                       'num_rating', 
                                                                                       'avg_rating']]
```

### Collaborative Filtering
- Similar book prediction based on users feedback

```python
books_with_ratings.head()
```

```python
# Grouping base on user-id tells the no.of books rated by each user:
x = books_with_ratings.groupby('User-ID').count()
x
```

```python
x.index
```

```python
x.shape
```

```python
# Select only users who atleast gave feedback for 200 books (Power Users)
x = x['Book-Rating'] > 200
x
```

```python
power_users = x[x].index
```

```python
power_users
```

```python
power_users.shape
```

```python
power_users.sort_values()
```

```python
# Selecting only records of power users
filtered_ratings = books_with_ratings[books_with_ratings['User-ID'].isin(power_users)]
```

```python
filtered_ratings
```

```python
# I am considering only the best users (given feedback on atleast 200 books) and group them based on the group title
y = filtered_ratings.groupby('Book-Title').count()
```

```python
y
```

```python
# The above dataframe tell how many users have read the book.
y.sort_values('User-ID',ascending=False)
```

```python
y = y['User-ID'] >= 50
```

```python
y
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
# pivot table giving the ratings for each book from each user
# Book row with userId as column
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
# Null values should be filled with zeros
pt = pt.fillna(0)
```

```python
pt
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
type(similarity_scores)
```

```python
df_temp = pd.DataFrame(similarity_scores)
```

```python
df_temp
```

```python
pt.index
```

```python
df_temp_name = df_temp
```

```python
df_temp_name.index = pt.index
df_temp_name.columns = pt.index
```

```python
# This gives the similarity scores between books
df_temp_name
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x: x[1], reverse=True)[1:6]
    # Lets create empty list and in that list I want to populate with the book information
    # Book Author book-title image url
    # Empty list

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
        
```

```python
# On the basis of 1984 code recommended similar books to 1984
recommend('1984')
```

```python
recommend('Animal Farm')
```
